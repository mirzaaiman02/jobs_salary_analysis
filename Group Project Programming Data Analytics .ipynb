Data preparation

```python
# to check for missing values

missing_values = jobs_salary.isnull().sum()
missing_values
```

```python
jobs_salary.info()
```

```python
# Verify unique values in categorical columns

for col in ['job_title', 'job_category', 'salary_currency', 'employee_residence', 'experience_level', 'employment_type', 'work_setting', 'company_location', 'company_size']:
    unique_values = jobs_salary[col].unique()
    print(f'Unique values in {col}: {unique_values}\n')
```

```python
# Standardize categorical variables
jobs_salary['experience_level'] = jobs_salary['experience_level'].str.upper()
jobs_salary['employment_type'] = jobs_salary['employment_type'].str.upper()
jobs_salary['work_setting'] = jobs_salary['work_setting'].str.upper()
jobs_salary['company_size'] = jobs_salary['company_size'].str.upper()
```

```python
## to see the data types for each columns
jobs_salary.dtypes
```

```python
# To clean the data
print(jobs_salary.isnull().sum())
jobs_salary = jobs_salary.drop_duplicates()
```

```python
## This visualize to show the top-paying job title in data science
plt.figure(figsize=(14, 8))
top_job_titles = jobs_salary.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_job_titles.values, y=top_job_titles.index)
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Title')
plt.title('Top 10 Highest Paying Job Titles in Data Science (per year)')
plt.show()
```

Findings for job title vs average salary in usd

This visualize is use horizontal bar chart to display the Top 10 highest paying job title in data science:

- The visualization employs a horizontal bar chart to highlight the top 10 highest-paying job titles in data science.
- It reveals that the "Analytics Engineering Manager" job title claims the highest average salary in data science, earning $400,000 USD annually.
- Following closely is the role of "Data Science Tech Lead," which secures the second-highest average salary at $350,000 per year.
- Conversely, the job title "Data Analytics Lead" emerges as the lowest-paying among the top 10, with an average salary of USD 200,000 per year.

```python
# this visualize is to show salary trends over the year by experience level
plt.figure(figsize=(12, 6))
sns.lineplot(x='work_year', y='salary_in_usd', hue='experience_level', data=jobs_salary, marker='o')
plt.title('Salary Trends Over the Years by Experience Level')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.legend(title='Experience Level')

# Ensure each year is displayed on the x-axis
years = jobs_salary['work_year'].unique()
plt.xticks(ticks=years, labels=years)

plt.show()
```

Findings Salary trends over the years by experience level vs average salary in USD

This visualization objective is to show salary trends over the years across different experience levels in the data science field.

Based on this line chart: 

- When the experience level becomes highest, such as executive level, the average salary in USD (per year) increases by more than $15,000.
- When the experience level becomes lowest, such as entry-level, the average salary in USD (per year) decreases to around $50,000 and above.
- Across all experience levels, there appears to be a general upward trend in salaries over the years.
- While there is an overall increasing trend, there are slight fluctuations in salaries from year to year.

Experience level impact:

Entry level:
Entry-level professionals typically start with lower salaries, gradually increasing as they gain experience.

Mid-Level: 
Salaries for mid-level professionals show steady growth over the years.

Senior level and Executive level:
Senior-level and Executive professionals command higher salaries, with notable increases over time, indicating the value placed on their extensive experience and leadership.

```python
# Salary distribution across job category

plt.figure(figsize=(12, 8))
category_salary = jobs_salary.groupby('job_category')['salary_in_usd'].mean().sort_values()
sns.barplot(x=category_salary.values, y=category_salary.index, palette='viridis')
plt.title('Average Salary by Job Category')
plt.xlabel('Average Salary in USD (per year)')
plt.ylabel('Job Category')
plt.show()
```

Findings Average Salary by job category vs average salary in USD (per year)

Based on this horizontal chart:

- The chart displays the average salary in USD (per year) starting from the lowest average salary to the highest average salary.
- The highest average salary in USD (per year) by job category is Machine Learning and AI, followed by Data Science and Research.
- The lowest average salary in USD (per year) by job category is Data Quality and Operations, which is around $100,000.
- Job seekers and students looking to find an internship can use this information to understand salary expectations across different job categories.

```python
## Heatmap of salary by experience level and employment type

salary_pivot = jobs_salary.pivot_table(values='salary_in_usd', index='employment_type', columns='experience_level', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(salary_pivot, annot=True, fmt=".0f", cmap='Blues')
plt.title('Heatmap of Salary by Experience Level and Employment Type')
plt.xlabel('Experience Level')
plt.ylabel('Employment Type')
plt.show()
```

Findings Heatmap of Salary by Experience Level and Employment Type

The objectives of this heatmap chart:

The heatmap visualizes the average salary distribution across different combinations of experience levels and employment types in the data science field.

- The heatmap showcases variations in average salaries based on both experience level and employment type.
- Certain combinations of experience level and employment type yield higher average salaries compared to others.
- We can spot patterns where specific experience levels or employment types consistently lead to higher or lower salaries.
- This visualization helps us understand how salaries vary based on both experience level and employment type.
- Employers can compare their salary offerings with industry standards for different experience levels and employment types.
- Job seekers can use this to plan their careers and choose roles with higher earning potential based on their experience level and preferred employment type.
